```python
# Your code goes here
import pandas as pd

url1 = r"C:\Users\abarr\Desktop\ironhack\week2\labs\insurance_customer.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
df1.shape
```

```python
df2.columns = [col.lower() for col in df2.columns]
df2.columns = [col.replace(" ", "_") for col in df2.columns]
df2.columns = ["state" if col == "st" else col for col in df2.columns ]
# df2.columns
```

```python
df2.shape
```

```python
df3.columns = [col.lower() for col in df3.columns]
df3.columns = [col.replace(" ", "_") for col in df3.columns]
df3.columns = ["state" if col == "st" else col for col in df3.columns]
```

```python
df3.shape
```

```python
# now all dfX.columns are similar
df1.columns == df2.columns
```

```python
# an outer join through the horizontal axis allows to keep all data from 3 dataframes into a single one
df_all = pd.concat([df1, df2, df3], axis=0, join="outer")
```

```python
df_all.nunique()
```

```python
df_all["state"].unique()
replace_state = {"AZ" : "Arizona"}
df_all["state"].replace(replace_state, inplace = True)
df_all["state"].unique()
```

```python
df_all["gender"].unique()
replace_gender = {"female":"F", "Male":"M"}
df_all["gender"].replace(replace_gender, inplace = True)
df_all["gender"].unique()
```

```python
replace_car = {"Luxury Car":"Luxury", "Luxury SUV":"Luxury", "Sports Car":"Luxury"}
df_all["vehicle_class"].replace(replace_car, inplace = True)
df_all["vehicle_class"].unique()
```

```python
df_all["education"].replace(to_replace ="Bachelors", value = "Bachelor", inplace = True)
df_all["education"].unique()
```

```python
df_all["customer_lifetime_value"] = df_all["customer_lifetime_value"].str.replace("%","")
df_all["customer_lifetime_value"].str.split("/")
# df_all["customer_lifetime_value"].fillna()
# df_all["customer_lifetime_value"].apply(lambda row:row[1])
# look for lambda functions
```
